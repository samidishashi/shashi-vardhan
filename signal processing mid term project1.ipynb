#### Superpositioning four sinusoidal signal

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
fig.set_size_inches((10,10))
ts = 12e-6
fs = 1/ts
t = np.arange(0,15000/fs , 1.0/fs)
f0 = 1000
f1 = 2000
f2 = 3000
f3 = 5000
phi = np.pi/2
A1 = 50
A2 = 100
A3 = 150
A4 = 40
x1 = A1 * np.sin(2 * np.pi * f0 * t + phi)
x2 = A2 * np.sin(2 * np.pi * f1 * t + phi)
x3 = A3 * np.sin(2 * np.pi * f2 * t + phi)
x4 = A4 * np.sin(2 * np.pi * f3 * t + phi)
x = x1 + x2 + x3 + x4
plt.plot(t, x)
plt.axis([0, 0.009, -400, 400])
plt.xlabel('Time (t) µs ', size = 14)
plt.ylabel('Amplitude (A)', size = 14)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.title('Superposition of four sinusoidal signals', size = 14)
plt.show()
```

#### Taking the DFT of the signal 

```python
fig=plt.figure()
fig.set_size_inches((10,10))
L = len(x)
f = np.arange(0,fs/2,fs/15000)
Y = np.fft.fft(x)
P = abs(Y)/L
plt.subplot(211)
plt.plot(f,P[:7500])
plt.title('DFT of sinusoidal singal ', size = 14)
plt.xlabel('Frequency Hz ', size = 14)
plt.ylabel('Magnitude', size = 14)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
inx = np.where(f>5.1e3)
inx = inx[0]
plt.subplot(212)
plt.plot(f[:inx[0]], P[:inx[0]])
plt.title('DFT singal zoomed in for 5.1 MHz', size = 14)
plt.xlabel('Frequency Hz ', size = 14)
plt.ylabel('Magnitude', size = 14)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.subplots_adjust( hspace = .5 )
```

Spectral peaks can be seen at the frequencies of the sinusoidal signals which are superpositioned

ts(sampling time) = 12µs

fs(samping frequency) = 1/ts

Equation for the sinusoidal signal :

$$ y(t) = Asin(2\pi f t + φ) $$

DFT is defined as :
$$ A_k=\sum_{m=0}^{n-1}a_m \exp \left(-2\pi i \frac{mk}{n} \right ) [k=0,...,n-1] $$

#### Adding a noise to the signal

```python
fig=plt.figure()
fig.set_size_inches((10,10))
f4 = 4e50
noise =100* np.cos(2 * np.pi * f4 * t + np.pi)
xnoise = x + noise
plt.plot(t,xnoise)
plt.axis([0, 0.009, -400, 500])
plt.xlabel('Time (t) µs ', size = 14)
plt.ylabel('Amplitude (A)', size = 14)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.title('Adding noise to the superposed signal', size = 14)
plt.show()
```

#### DFT of the noisy signal

```python
fig=plt.figure()
fig.set_size_inches((10,10))
Lnoise = len(xnoise)
fnoise = np.arange(0,fs/2,fs/15000)
Ynoise = np.fft.fft(xnoise)
Pnoise = abs(Ynoise)/Lnoise
plt.subplot(211)
plt.plot(fnoise,Pnoise[:7500])
plt.title('DFT of noisy sinusoidal singal ', size = 14)
plt.xlabel('Frequency Hz ', size = 14)
plt.ylabel('Magnitude', size = 14)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
inxnoise = np.where(fnoise>5.1e3)
inxnoise = inxnoise[0]
plt.subplot(212)
plt.plot(fnoise[:inx[0]], Pnoise[:inxnoise[0]])
plt.title('DFT singal zoomed in for 5.1 MHz', size = 14)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.xlabel('Frequency Hz ', size = 14)
plt.ylabel('Magnitude', size = 14)
plt.subplots_adjust( hspace = .5 )
```

When the noise is high the shape of the signal cannot be identitfied in the time domain.

Number of peaks increases in the frequency domain and the hihgest peak is at the frequency of the noise signal.



#### Filtering out the 2MHz peak from the signal

```python
from __future__ import division
from scipy.signal import butter, lfilter


cutoff = 2e3 
b, a = butter(5, cutoff / (fs / 2), btype='high') 
filtered_signal = lfilter(b, a, x, axis=0)
plt.xlabel('Time ', size = 14)
plt.ylabel('Amplitude', size = 14)
plt.title('Filtering out the 2MHz peak', size=14)
plt.axis([0, 0.0125, -300, 300])
plt.plot(t,filtered_signal)
```

Used a Butterworth filter and IIR filter to get cutoff the frequency of the 2MHz peak from the signal.

5 is the order of the filter.

frequency to be eliminated is 2MHz.


#### Frequency response of the Butterworth filter


```python
from scipy import signal
import matplotlib.pyplot as plt

b, a = signal.butter(4, 2000, 'high', analog=True)
w, h = signal.freqs(b, a)
plt.plot(w, 20 * np.log10(abs(h)))
plt.xscale('log')
plt.title('Butterworth filter frequency response', size=14)
plt.xlabel('Frequency [radians / second]',size=14)
plt.ylabel('Amplitude [dB]', size=14)
plt.margins(0, 0.01)
plt.grid(which='both', axis='both')
plt.axvline(2000, color='green')
plt.show()
```
